Now that we have the pandas dataframe, lets experiment with getting the number of passengers who survived.

```python
total_passengers = len(data["PassengerId"])
print("Total Passengers: %s" % total_passengers)

total_survived = np.sum(data["Survived"])
print("Total Survived: %s" % total_survived)
```

## Find the male and female stats

### Female Stats

```python
women_only_stats = data.loc[data["Sex"] == "female"]
women_only_stats
```

### Male Stats

```python
men_only_stats = data.loc[data["Sex"] == "male"]
men_only_stats
```

## Proportions

Now that we have the proportions, we can use that to figure out how many females and males died in the incident.

```python
total_women = len(women_only_stats)
women_survived = len(women_only_stats.loc[data["Survived"] == 1])
proportion_women_survived = women_survived / total_women
print("Ratio Women Survived: %s" % proportion_women_survived)
```

```python
total_men = len(men_only_stats)
men_survived = len(men_only_stats.loc[data["Survived"] == 1])
proportion_men_survived = men_survived / total_men
print("Ratio Men Survived: %s" % proportion_men_survived)
```
